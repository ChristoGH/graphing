```python
row=CLM_GB_CUSTOMERS_BASE_REPORT_df.iloc[0]
dedupestatic=np.format_float_scientific(row['MASTER_CIS'], unique=False, precision=11, trim='.')
def neo4j_query_fn(row):
    dedupestatic=row['MASTER_CIS']
    CLM_ENROL_DATE=row['CLM_ENROL_DATE']
    ENROLMENT_CHANNEL=row['ENROLMENT_CHANNEL']
    STATUS=row['STATUS']
    IREWARDS_NO=row['IREWARDS_NO']
    CLM_REWARDS_NUMBER=row['CLM_REWARDS_NUMBER']
    AFFINITY=row['AFFINITY']
    GB_APP_DOWNLOAD=row['GB_APP_DOWNLOAD']
    MONEY_APP_DOWNLOAD=row['MONEY_APP_DOWNLOAD']
    CARD_SWIPER_ELIGIBLE=row['CARD_SWIPER_ELIGIBLE']
    CARD_SWIPER_OPTED=row['CARD_SWIPER_OPTED']
    CARD_SWIPER_OPTED_DATE=row['CARD_SWIPER_OPTED_DATE']
    # -----------------------------------------------------    
    MONEY_MANAGER_ELIGIBLE=row['MONEY_MANAGER_ELIGIBLE']
    MONEY_MANAGER_OPTED=row['MONEY_MNAGER_OPTED']
    MONEY_MANAGER_OPTED_DATE=row['MONEY_MNAGER_OPTED_DATE']
    RESPONSIBLE_BORROWER_ELIGIBLE=row['RESPONSIBLE_BORROWER_ELIGIBLE']
    RESPONSIBLE_BORROWER_OPTED=row['RESPONSIBLE_BORROWER_OPTED']
    RESPONSIBLE_BORROWER_OPTED_DATE=row['RESPONSIBLE_BORROWER_OPTED_DATE']
    GREENBACKS=row['GREENBACKS']
    PROBABILISTIC=row['PROBABILISTIC']
    CAMPAIGN_POINTS=row['CAMPAIGN_POINTS']
    GREENBACKS_RANDVALUE=row['GREENBACKS_RANDVALUE']
    # -----------------------------------------------------    
    PROBABILISTIC_RANDVALUE=row['PROBABILISTIC_RANDVALUE']
    GREENBACKS_RANDRANGE=row['GREENBACKS_RANDRANGE']
    PROBABILISTIC_RANDRANGE=row['PROBABILISTIC_RANDRANGE']

    return """MATCH (client:GreenBacks) WHERE client.dedupestatic={dedupestatic}
    SET client.CLM_ENROL_DATE = coalesce(date('{CLM_ENROL_DATE}'),"not applicable"),
    client.ENROLMENT_CHANNEL = coalesce('{ENROLMENT_CHANNEL}',"not applicable"),
    client.STATUS = coalesce('{STATUS}',"not applicable"),
    client.IREWARDS_NO = coalesce(toInteger('{IREWARDS_NO}'),"not applicable"),
    client.CLM_REWARDS_NUMBER = coalesce(toInteger('{CLM_REWARDS_NUMBER}'),"not applicable"),
    client.AFFINITY = coalesce('{AFFINITY}',"not applicable"),

    client.GB_APP_DOWNLOAD = coalesce(toInteger('{GB_APP_DOWNLOAD}'),"not applicable"),
    client.MONEY_APP_DOWNLOAD = coalesce(toInteger('{MONEY_APP_DOWNLOAD}'),"not applicable"),

    client.CARD_SWIPER_ELIGIBLE = coalesce(toInteger('{CARD_SWIPER_ELIGIBLE}'),"not applicable"),
    client.CARD_SWIPER_OPTED = coalesce(toInteger('{CARD_SWIPER_OPTED}'),"not applicable"),
    client.CARD_SWIPER_OPTED_DATE=coalesce(date('{CARD_SWIPER_OPTED_DATE}'),"not applicable"),

    client.MONEY_MANAGER_ELIGIBLE=coalesce(toInteger('{MONEY_MANAGER_ELIGIBLE}'),"not applicable"),
    client.MONEY_MANAGER_OPTED=coalesce(toInteger('{MONEY_MANAGER_OPTED}'),"not applicable"),
    client.MONEY_MANAGER_OPTED_DATE = coalesce(date('{MONEY_MANAGER_OPTED_DATE}'),"not applicable"),

    client.RESPONSIBLE_BORROWER_ELIGIBLE = coalesce(toInteger('{RESPONSIBLE_BORROWER_ELIGIBLE}'),"not applicable"),
    client.RESPONSIBLE_BORROWER_OPTED = coalesce(toInteger('{RESPONSIBLE_BORROWER_OPTED}'),"not applicable"),
    client.RESPONSIBLE_BORROWER_OPTED_DATE = coalesce(date('{RESPONSIBLE_BORROWER_OPTED_DATE}'),"not applicable"),

    client.GREENBACKS = coalesce(toInteger('{GREENBACKS}'),"not applicable"),
    client.PROBABILISTIC = coalesce(toInteger('{PROBABILISTIC}'),"not applicable"),
    client.CAMPAIGN_POINTS = coalesce(toInteger('{CAMPAIGN_POINTS}'),"not applicable"),
    client.GREENBACKS_RANDVALUE = coalesce(toFloat('{GREENBACKS_RANDVALUE}'),"not applicable"),

    client.PROBABILISTIC_RANDVALUE = coalesce(toFloat('{PROBABILISTIC_RANDVALUE}'),"not applicable"),
    client.GREENBACKS_RANDRANGE = coalesce('{GREENBACKS_RANDRANGE}',"not applicable"),
    client.PROBABILISTIC_RANDRANGE = coalesce('{PROBABILISTIC_RANDRANGE}',"not applicable")
    RETURN count(client) AS client_count;

    """.format(dedupestatic=dedupestatic,
                               CLM_ENROL_DATE=CLM_ENROL_DATE,
                                ENROLMENT_CHANNEL=ENROLMENT_CHANNEL,
                                STATUS=STATUS,
               IREWARDS_NO=IREWARDS_NO,
                                CLM_REWARDS_NUMBER=CLM_REWARDS_NUMBER,
                                AFFINITY=AFFINITY,
                                GB_APP_DOWNLOAD=GB_APP_DOWNLOAD,
                                MONEY_APP_DOWNLOAD=MONEY_APP_DOWNLOAD,
                                CARD_SWIPER_ELIGIBLE=CARD_SWIPER_ELIGIBLE,
                                CARD_SWIPER_OPTED=CARD_SWIPER_OPTED,
                                CARD_SWIPER_OPTED_DATE=CARD_SWIPER_OPTED_DATE,
                            # -----------------------------------------------------    
                                MONEY_MANAGER_ELIGIBLE=MONEY_MANAGER_ELIGIBLE,
                                MONEY_MANAGER_OPTED=MONEY_MANAGER_OPTED,
                                MONEY_MANAGER_OPTED_DATE=MONEY_MANAGER_OPTED_DATE,
                                RESPONSIBLE_BORROWER_ELIGIBLE=RESPONSIBLE_BORROWER_ELIGIBLE,
                                RESPONSIBLE_BORROWER_OPTED=RESPONSIBLE_BORROWER_OPTED,
                                RESPONSIBLE_BORROWER_OPTED_DATE=RESPONSIBLE_BORROWER_OPTED_DATE,
                                GREENBACKS=GREENBACKS,
                                PROBABILISTIC=PROBABILISTIC,
                                CAMPAIGN_POINTS=CAMPAIGN_POINTS,
                                GREENBACKS_RANDVALUE=GREENBACKS_RANDVALUE,
                            # -----------------------------------------------------    
                                PROBABILISTIC_RANDVALUE=PROBABILISTIC_RANDVALUE,
                                GREENBACKS_RANDRANGE=GREENBACKS_RANDRANGE,
                                PROBABILISTIC_RANDRANGE=PROBABILISTIC_RANDRANGE)
```
